```python
import pandas as pd
```

```python
df = pd.DataFrame(
    {
        "Name": [
            "Braund, Mr. Owen Harris",
            "Allen, Mr. William Henry",
            "Bonnell, Miss. Elizabeth"
        ],
        "Age": [22, 35, 58],
        "Sex": ["male", "male", "female"]
    }
)
print(df.describe())
print()
print(df.describe(include="object"))
print()
print(df.describe(include="all"))
```

```python
print(df["Age"])
```

```python
ages = pd.Series([22, 35, 58], name="Age")
print(ages)
print(type(ages))
```

```python
print(df["Age"].max()) # DataFrame
print(ages.max()) # Series
```

```python
ages = titanic["Age"]
print(ages)
print(f"type: {type(ages)}")
print(ages.shape)
```

```python
print(titanic[["Age", "Sex"]])
print(type(titanic[["Age", "Sex"]]))
print(titanic[["Age", "Sex"]].shape)
```

```python
above_35 = titanic[titanic["Age"] > 35]
print(above_35)
```

```python
print(titanic["Age"] > 35)
```

```python
print(titanic["Pclass"].isin([2, 3]))
```

```python
class_23 = titanic[titanic["Pclass"].isin([2, 3])]
```

```python
ages_no_na = titanic[titanic["Age"].notna()]
print(ages_no_na)
```

```python
adult_names = titanic.loc[titanic["Age"] > 35, "Name"]
print(adult_names)
```
